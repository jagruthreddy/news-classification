```python
import numpy as np # linear algebra
import pandas as pd # data processing
import nltk

#import
from google.colab import drive
drive.mount('/content/gdrive')
dft = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/News Classification/True.csv") #True data
dff = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/News Classification/Fake.csv") #Fake data
```

```python
#This tokenizer divides a text into a list of sentences
#by using an unsupervised algorithm to build a model for abbreviation words, collocations, and words that start sentences.
nltk.download('punkt') 
```

```python
# we need 2 classifications 
dff['genuiness']=0
dft['genuiness']=1
```

```python
df = pd.concat([dff,dft])
df
```

PRE PROCESSING

```python
df = df.reset_index(drop=True)
df = df.drop(['date','title','subject'],axis=1)
df
```

```python
from nltk.tokenize import word_tokenize
df['text'] = df['text'].apply(word_tokenize)
```

```python
df
```

```python
from nltk.stem.snowball import SnowballStemmer
sb = SnowballStemmer('english')
```

```python
#Stemming amd lower cases 
def stem_it(text):
  return [sb.stem(word) for word in text]
```

```python
df['text'] = df['text'].apply(stem_it)
```

```python
df
```

```python
def stopword_removal(text):
  return [word for word in text if len(word)>>2]
```

```python
df['text'] = df['text'].apply(stopword_removal)
```

```python
df
```

```python
df['text'] = df['text'].apply(' '.join)
```

```python
df
```

MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
x= df['text']
y= df['genuiness']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
```

```python
tfidfv = TfidfVectorizer(max_df=0.7)
```

```python
tfidfv_train = tfidfv.fit_transform(x_train)
tfidfv_test = tfidfv.transform(x_test)
```

```python
lr= LogisticRegression(max_iter =500)
lr.fit(tfidfv_train, y_train)
```

```python
y_pred = lr.predict(tfidfv_test)
accuracy_score(y_test, y_pred)
```

```python
#Comparison
from sklearn.linear_model import PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter =100)
pac.fit(tfidfv_train, y_train)
```

```python
y_pred_pac = pac.predict(tfidfv_test)
accuracy_score(y_test, y_pred_pac)
```
